```python
import tensorflow as tensorflow
from tensorflow.keras import models, layers
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
IMAGE_SIZE = 256
BATCH_SIZE = 32
dataset = tensorflow.keras.preprocessing.image_dataset_from_directory(
    "./PlantVillage/",
    shuffle=True,
    image_size= (IMAGE_SIZE , IMAGE_SIZE),
    batch_size= BATCH_SIZE
)
```

```python
class_names = dataset.class_names
```

```python
for image_batch , label_batch in dataset.take(1):
    plt.imshow(image_batch[0].numpy().astype("uint8"))
    plt.title(class_names[label_batch[0]])
```

```python
train_ds = dataset.take(54)
test_ds = dataset.skip(54)
val_ds = test_ds.skip(7)
test_ds = test_ds.take(7)
```

```python
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=tensorflow.data.AUTOTUNE)
val_ds = val_ds.cache().shuffle(1000).prefetch(buffer_size=tensorflow.data.AUTOTUNE)
test_ds = test_ds.cache().shuffle(1000).prefetch(buffer_size=tensorflow.data.AUTOTUNE)
```

```python
resize_and_rescale = tensorflow.keras.Sequential([
    layers.experimental.preprocessing.Resizing(IMAGE_SIZE , IMAGE_SIZE),
    layers.experimental.preprocessing.Rescaling(1.0/255)
])
```

```python
data_augumentation = tensorflow.keras.Sequential([
    layers.experimental.preprocessing.RandomFlip('horizontal_and_vertical'),
    layers.experimental.preprocessing.RandomRotation(0.2)
])
```

```python
n_classes = len(class_names)
```

```python
model = models.Sequential([
    resize_and_rescale,
    data_augumentation,
    layers.Conv2D(
        32,
        (3,3),
        activation='relu',
        input_shape=(BATCH_SIZE,IMAGE_SIZE , IMAGE_SIZE, 3),
    ),
    layers.MaxPool2D(2,2),
    layers.Conv2D(
        64,
        (3,3),
        activation='relu',
    ),
    layers.MaxPool2D(2,2),
    layers.Conv2D(
        64,
        (3,3),
        activation='relu',
    ),
    layers.MaxPool2D(2,2),
    layers.Conv2D(
        64,
        (3,3),
        activation='relu',
    ),
    layers.MaxPool2D(2,2),
    layers.Conv2D(
        64,
        (3,3),
        activation='relu',
    ),
    layers.MaxPool2D(2,2),
    layers.Conv2D(
        64,
        (3,3),
        activation='relu',
    ),
    layers.MaxPool2D(2,2),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(
        n_classes,
        activation='softmax'
    )
    
])
model.build(input_shape=(BATCH_SIZE , IMAGE_SIZE, IMAGE_SIZE , 3))
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss=tensorflow.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    metrics=['accuracy']
)
```

```python
history = model.fit(
    train_ds,
    epochs=30,
    batch_size=BATCH_SIZE,
    verbose=1,
    validation_data=val_ds,
)
```
